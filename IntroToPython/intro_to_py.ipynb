## 2. "Hello World" & More Syntax

```python
# The simplest program
print("Hello, World!")
```

```python
# Variables — no type declarations needed
name    = "Alice"
age     = 17
gpa     = 3.95
is_cool = True

print(f"{name} is {age} years old")
print(f"Type of age: {type(age)}")
print(f"Type of gpa: {type(gpa)}")
```

```python
# Conditionals — indentation (TAB) defines blocks, no curly braces {}
score = 85

if score >= 90:
    print("A")
elif score >= 80:
    print("B")
else:
    print("C")

# one-line conditional
grade = "pass" if score >= 60 else "fail"
print(grade)
```

```python
# Loops
for i in range(5):
    print(i, end=" ")
print()

# Loop over a list
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
```

```python
# Functions
def greet(name, greeting="Hello"):
    return f"{greeting}, {name}!"

print(greet("Bob"))
print(greet("Bob", greeting="Hey"))

# Functions are first-class, meaning you can pass them around
def apply(func, value):
    return func(value)

print(apply(str.upper, "hello"))
```

### Some Advanced Python

```python
# Lambdas —  one-liner functions
square = lambda x: x ** 2
print(square(5))

# map, filter, sorted with lambdas
nums = [3, 1, 4, 1, 5, 9, 2, 6]
print(list(map(lambda x: x**2, nums)))
print(list(filter(lambda x: x > 3, nums)))
print(sorted(nums, key=lambda x: -x))   # sort descending
```

```python
#My version

import time

x = [100, 59, 490, 1, 18, 0, 3, 2, 1, 4, 5, 6, 7, 8, 9]
y = [50, 70, 90, 110, 130, 150, 170, 190, 210, 230, 250, 270, 290, 310, 330]

def main(x, y):
    start=time.time()
    prod = list(map(lambda a,b: a*b, x, y))
    for i in range(0,15):
        print(prod[i], end=" ")
    print()
    stop=time.time()
    print(f"Execution time: ",stop-start)
    pass


main(x, y)
```
